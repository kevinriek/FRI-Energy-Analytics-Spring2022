```python
import numpy as np                        # ndarrys for gridded data
import pandas as pd                       # DataFrames for tabular data
import os                                 # set working directory, run executables
import matplotlib.pyplot as plt           # for plotting
import matplotlib.patches as mpatches
import copy                               # for deep copies
from sklearn.metrics import mean_squared_error as mse
from sklearn.preprocessing import MinMaxScaler
```

```python
#Display all of the rows in a dataframe
pd.set_option('display.max_rows', 600)
```

```python
#Import datasets from .xslx as pandas DataFrames
filename = os.path.abspath('') + '\\Data\\Filtered\\FilteredProduction.xlsx'
file = open(filename, "rb")
prodDF = pd.read_excel(file, engine="openpyxl")

filename = os.path.abspath('') + '\\Data\\Filtered\\FilteredInjection.xlsx'
file = open(filename, "rb")
injectDF = pd.read_excel(file, engine="openpyxl")
```

```python
#group datasets into wells
p_wells = prodDF.groupby(prodDF.Well_Name)
i_wells = injectDF.groupby(['Enhanced_Recovery_Unit', 'Well_Name'])

#Option to only examine the wells of a specific Recovery Unit
oneUnit = False
unitName = "Buffalo Red River Unit"

#If option on, restrict dataframes to only the wells of a given unit
if (oneUnit):
    p_wells = prodDF.groupby(prodDF.Enhanced_Recovery_Unit)
    unit = p_wells.get_group(unitName)
    p_wells = unit.groupby(unit.Well_Name)
    
    i_wells = injectDF.groupby(injectDF.Enhanced_Recovery_Unit)
    unit = i_wells.get_group(unitName)
    i_wells = unit.groupby(unit.Well_Name)
```

```python
#Summary Statistics for Production Wells
Well, numDates, oilMean, start = [], [], [], []
waterMean, meanDays, Unit = [], [], []

for well, data in p_wells:
    Well.append(well)
    oilMean.append(data["Oil_bbls"].mean())
    waterMean.append(data["Water_bbls"].mean())
    numDates.append(len(pd.unique(data["Date_Production"])))
    meanDays.append(data["Prod_Days"].mean())
    start.append(data.iloc[0].Date_Production)
    units = pd.unique(data["Enhanced_Recovery_Unit"])

    if ((len(units) > 1) & (units[0] == "None")):
        Unit.append("[" + units[1] + "]")
    else:
        Unit.append(units)


data = pd.DataFrame({'Production Well': Well, 'Unit': Unit, 'OilMean': oilMean, 
                   'WaterMean':waterMean, 'MeanDays':meanDays, 'NumDates': numDates, 'StartDate':start})

display(data)
```

```python
#Summary Statistics for Injection Wells
Well, wellCount, Unit, Type = [], [], [], []
meanDays, Rate, volMean, numDates =  [], [], [], []

for well, data in i_wells:
    Well.append(well)
    wellCount.append(len(pd.unique(data["Well_Name"])))
    Type.append(pd.unique(data["Type"]))
    meanDays.append(data["Days_Injection"].mean()) 
    Rate.append(data["InjectRate_Daily_Avg_Calc"].mean())    
    volMean.append(data["Vol_Fluid_Injected"].mean()) 
    numDates.append(len(pd.unique(data["Date_Injection_Report"])))
    units = pd.unique(data["Enhanced_Recovery_Unit"])

    if ((len(units) > 1) & (units[0] == "None")):
        Unit.append("[" + units[1] + "]")
    else:
        Unit.append(units)
    

data = pd.DataFrame({'Well, Unit': Well, 'InjectionType':Type, 'Unit':Unit, 'NumDates':numDates,
                     'MeanDays':meanDays, 'InjectRate':Rate, 'VolMean':volMean})

display(data)
```

```python
#Group datasets into enhanced recovery units
p_units = prodDF.groupby(prodDF.Enhanced_Recovery_Unit)
i_units = injectDF.groupby(injectDF.Enhanced_Recovery_Unit)
```

```python
#Summary Statistics for Recovery Units
Unit, pCount, pDates, pMean = [], [], [], []
iCount, iDates, iType, iMean = [], [], [], []

for unit, p_unit in p_units:
    Unit.append(unit)
    pCount.append(len(pd.unique(p_unit["Well_Name"])))
    pMean.append(p_unit["Oil_bbls"].mean())
    pDates.append(len(pd.unique(p_unit["Date_Production"])))
    
    if (unit in i_units.groups):
        i_unit = i_units.get_group(unit)
        iCount.append(len(pd.unique(i_unit["Well_Name"])))
        iDates.append(len(pd.unique(i_unit["Date_Injection_Report"])))
        iType.append(pd.unique(i_unit["Type"]))
        iMean.append(i_unit["Vol_Fluid_Injected"].mean())
    else:
        iCount.append(0)
        iDates.append(0)
        iType.append("[None]")
        iMean.append(0)

data = pd.DataFrame({'Unit': Unit, 'ProdWells':pCount, 'ProdDates':pDates, 'ProdMean':pMean,
                     'InjWells':iCount, 'InjDates':iDates, 'InjectionType':iType, 'InjMean':iMean})

display(data)
```

```python
#Plots all of the wells in a given group in separate graphs
def plot_groups(wells, type):
    for well, data in wells:
        Well = wells.get_group(well)
        if (type == "Production"):
            df = pd.DataFrame({'date':Well.Date_Production,'production':Well.Oil_bbls})
        elif (type == "Injection"):
            df = pd.DataFrame({'date':Well.Date_Injection_Report,'injection':Well.Vol_Fluid_Injected})
        else:
            raise Exception("Invalid Well Type")
        df = df.set_index(['date'])
            
        plt.title(well + " " + type)
        plt.plot(df)
        plt.subplots_adjust(left=0.0, bottom=0.0, right=1.5, 
                        top=1.0, wspace=0.2, hspace=0.2)
            
        plt.xlabel("Date")
        plt.ylabel("Bbls")
        plt.figure()
```

```python
#Plots all of the wells in a given enhanced recovery unit
#including injection wells. Option to scale all data. 
def plot_unit(unit_name, scaledOn):
    
    p_Unit = p_units.get_group(unit_name)
    i_Unit = i_units.get_group(unit_name)
    
    pMax = 1
    iMax = 1
    if(scaledOn):
        pMax = p_Unit.Oil_bbls.max()
        iMax = i_Unit.Vol_Fluid_Injected.max()
    
    pUnit_wells = p_Unit.groupby(p_Unit.Well_Name)
    iUnit_wells = i_Unit.groupby(i_Unit.Well_Name)
    
    plots = []
    
    for well, data in pUnit_wells:
        Well = pUnit_wells.get_group(well)
        df = pd.DataFrame({'date':Well.Date_Production ,'production':(Well.Oil_bbls/pMax)})
        df = df.set_index(['date'])
        plt.plot(df, color='r')
        
    for well, data in iUnit_wells:
        Well = iUnit_wells.get_group(well)
        df = pd.DataFrame({'date':Well.Date_Injection_Report,'injection':(Well.Vol_Fluid_Injected/iMax)})
        df = df.set_index(['date'])
        plt.plot(df, color='b')
        
    plt.title(unit_name)
    plt.subplots_adjust(left=0.0, bottom=0.0, right=2.5, 
                        top=1.5, wspace=0.2, hspace=0.2)
    
    prod_leg = mpatches.Patch(color='red', label='Production')
    inj_leg = mpatches.Patch(color='blue', label='Injection')
    plt.legend(handles = [prod_leg, inj_leg], loc="upper right")
    
    plt.xlabel("Date")
    if (scaledOn):
        plt.ylabel("Scaled Bbls")
    else:
        plt.ylabel("Bbls")
    plt.figure()
```
